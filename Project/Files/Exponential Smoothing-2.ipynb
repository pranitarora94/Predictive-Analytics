```r
library(forecast)
library(caret)
```

```r
training<-read.csv("data/TrainingSet.csv");
testing<-read.csv("data/SubmissionRows.csv");
```

```r
#head(training)
```

```r
#extract values we are testing
testVals<-testing$X
training_data <- training[which(training$X%in%testVals),]; 
#modify columns and make them into time series 
#list of time series

#clean up years labels
colnames(training_data)[2:37] <-c(1972:2007);
#colnames(training_data)[2:37]
#training_data

#reformat 
t_df<-data.frame(A= numeric(0), B= numeric(0))

ts_list<-c();
for(i in seq_len(nrow(training_data))){

    id<-training_data[i,]$X
    data = training_data[i,2:37]
    rownames(data)<-id
    t_df<-rbind(t_df,data)
    #make into time series
   
    
}

t_df
ts_list
```

```r
t_transform<-t(t_df)

predictions<-c();
#print( ncol(t_transform))
# for(i in seq_len(ncol(t_transform))){
#     print(i)
#     time_series<-ts(t_transform[,i],start=1972,freq=1)
#     time_series<-na.omit(time_series)
# }
for(i in seq_len(ncol(t_transform))){
    #print(i)
    #print(t_transform[,i])
    #convert to time series
    time_series<-ts(t_transform[,i],start=1972,freq=1)
    time_series<-ts(na.exclude(time_series))
    #perform exponential smooting
    if(i!=18 && i!=134 && i!=174 && i!=277 && i!=570 &&i!=680){
        sub<-head(time_series,-1)
        sub<-ts(sub) 
        fit <- HoltWinters(sub, beta=FALSE,gamma=FALSE)
        forecast<-forecast(fit, 1)
        mean<-forecast$mean[1]
        predictions <- rbind(predictions,mean)
    }
    else{
        predictions<-rbind(predictions,0)
    }
 
#     print(mean)
}




    
#     predictions <- rbind(predictions,mean)
    
```

```r
ref <- t_df["2007"]
nrow(predictions)

nrow(ref)

predictions[5];
ref[5,]



errTable<-function(references,predictions){
    errt<-data.frame(A= numeric(0), B= numeric(0))
    preds<-c(); 
    refs<-c();
    for(i in seq_len(nrow(predictions))){
        pred<-predictions[i];
        ref<-references[i,];
        preds<-rbind(preds,pred);
        refs<-rbind(refs,ref);
        
        
    }
    errt<-cbind(refs,preds); 
    
    return(errt)
}

errT<-errTable(ref,predictions);
colnames(errT)<-c("references","predictions"); 
rownames(errT)<-testVals;
errT
```

```r
# errT["references"]; 
#errT[,"references"];
# errT$references; 
expected <- errT[,1];
actual <- errT[,2]; 
error<- abs(actual - expected) / expected;  
#accuracy
errT<-cbind(errT,error)
errT
# errT$accuracy <- apply(subset(errT,select = c("references","predictions")),1,function(d){
#     print("hello");
#     1
# })
#z$wmean <- apply(subset(z, select = c(x, y)), 1, function(d) weighted.mean(d, weight, na.rm = TRUE));
```
