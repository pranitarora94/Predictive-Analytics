```r
library(forecast)
library(caret)
```

```r
training<-read.csv("data/TrainingSet.csv");
testing<-read.csv("data/SubmissionRows.csv");
```

```r
training
```

```r
#extract values we are testing
testVals<-testing$X
training_data <- training[which(training$X%in%testVals),]; 
#modify columns and make them into time series 
#list of time series

#clean up years labels
colnames(training_data)[2:37] <-c(1972:2007);
#colnames(training_data)[2:37]
#training_data

#reformat 
t_df<-data.frame(A= numeric(0), B= numeric(0))

ts_list<-c();
for(i in seq_len(nrow(training_data))){

    id<-training_data[i,]$X
    data = training_data[i,2:37]
    rownames(data)<-id
    t_df<-rbind(t_df,data)
    #make into time series
   
    
}

t_df
ts_list
```

```r
t_transform<-t(t_df);

t_transform
```

```r
predictions<-c(); 
for(i in seq_len(ncol(t_transform))){
    print(t_transform["2007",i])
    #get value for 2007 
#     time_series<-ts(t_transform[,i],start=1972,freq=1)
#     time_series<-ts(na.exclude(time_series))
#     print(time_series)
    
    
    
} 
```

```r

#print( ncol(t_transform))
# for(i in seq_len(ncol(t_transform))){
#     print(i)
#     time_series<-ts(t_transform[,i],start=1972,freq=1)
#     time_series<-na.omit(time_series)
# }

for(i in seq_len(ncol(t_transform))){
    
    #print(t_transform[,i])
    #convert to time series
    time_series<-ts(t_transform[,i],start=1972,freq=1)
    time_series<-ts(na.exclude(time_series))
    #perform exponential smooting
    if(i!=18 && i!=134 && i!=174 && i!=277 && i!=570 &&i!=680){
        fit <- HoltWinters(time_series,gamma=FALSE)
        forecast<-forecast(fit, 5)
        mean<-forecast$mean 
        predictions <- rbind(predictions,mean)
    }
    else{
        replace = t_transform["2007",i]; #added to see if prediction is better .... 
        predictions<-rbind(predictions,replace) ##changed from 0 to replace
    }
    
   
 
#     print(mean)
}



    
#     predictions <- rbind(predictions,mean)
    
```
